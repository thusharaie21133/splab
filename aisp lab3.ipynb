```python
import librosa
import soundfile as sf
import IPython.display as ipd
original_signal, sr = librosa.load(r"C:\Users\Hp\Documents\Sound Recordings\statement.wav", sr=None)
trimmed_signal, _ = librosa.effects.trim(original_signal)
sf.write('trimmed_audio.wav', trimmed_signal, sr)
print("Playing original audio...")
ipd.display(ipd.Audio(original_signal, rate=sr))
print("Playing trimmed audio...")
ipd.display(ipd.Audio(trimmed_signal, rate=sr))
```

```python
import librosa
import soundfile as sf
import IPython.display as ipd
original_signal, sr = librosa.load(r"C:\Users\Hp\Documents\Sound Recordings\statement.wav", sr=None)
top_db_values = [20, 30, 40, 50]
for top_db in top_db_values:
    intervals = librosa.effects.split(original_signal, top_db=top_db)
    split_signals = []
    for interval in intervals:
        split_signal = original_signal[interval[0]:interval[1]]
        split_signals.append(split_signal)
    print(f"Playing split audio with top_db={top_db}...")
    for idx, split_signal in enumerate(split_signals):
        print(f"Segment {idx + 1}: Duration {len(split_signal) / sr:.2f} seconds")
        ipd.display(ipd.Audio(split_signal, rate=sr))
```
