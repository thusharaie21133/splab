```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the original speech signal
file_path = r"C:\Users\Hp\Downloads\speech\THUSHAR.wav"
y, sr_original = librosa.load(file_path, sr=None)  # Load with the original sampling rate

# Plot the waveform of the original signal
plt.figure(figsize=(15, 5))
librosa.display.waveshow(y, sr=sr_original)
plt.title('Waveform of Original Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# Downsample the signal
target_sr = 8000  # Set the target sampling rate
y_downsampled = librosa.resample(y, orig_sr=sr_original, target_sr=target_sr)

# Plot the waveform of the downsampled signal
plt.figure(figsize=(15, 5))
librosa.display.waveshow(y_downsampled, sr=target_sr)
plt.title(f'Waveform of Downsampled Speech Signal (Target SR: {target_sr} Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# Listen to the original and downsampled signals
from IPython.display import Audio, display
display(Audio(y, rate=sr_original, normalize=False))  # Original signal
display(Audio(y_downsampled, rate=target_sr, normalize=False))  # Downsampled signal
```
